# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

# Exercise
1. Print path of cities visited
2. Take graph input from user
3. Test with different number of cities

```python
def print_path(path):
    print(" -> ".join(path))

# Example usage
cities = ["Delhi", "Agra", "Jaipur", "Udaipur"]
print_path(cities)
```

```python
def take_graph_input():
    graph = {}
    n = int(input("Enter number of cities: "))
    for i in range(n):
        city = input(f"Enter name of city {i+1}: ")
        connections = input(f"Enter connected cities for {city} (comma separated): ").split(",")
        graph[city] = [c.strip() for c in connections if c.strip()]
    return graph

# Example usage
graph = take_graph_input()
print("Graph:", graph)
```

```python
def test_graph():
    graph = take_graph_input()
    print("Graph created:", graph)
    path = list(graph.keys())  # simple path visiting all cities
    print("Path of cities visited:")
    print_path(path)

# Example run
test_graph()
```
